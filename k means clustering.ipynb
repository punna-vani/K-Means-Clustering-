# elbow method

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# TRAINING THE KMEANS MODEL OF DATA

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
plt.scatter(x[y_kmeans==0,0],x[y_means==0,1],s=100,c='orange',label='Cluster1')
plt.scatter(x[y_kmeans==1,0],x[y_means==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(x[y_kmeans==2,0],x[y_means==2,1],s=100,c='green',label='Cluster3')
plt.scatter(x[y_means==3,0],x[y_means==3,1],s=100,c='cyan',label='Cluster4')
plt.scatter(x[y_means==4,0],x[y_means==4,1],s=100,c='red',label='Cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='purple',label='centroids')
plt.title('Cluster of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score(1-100)')
plt.legend()
plt.show()
```
